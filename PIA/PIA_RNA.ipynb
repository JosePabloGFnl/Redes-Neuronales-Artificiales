```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook

%matplotlib inline

#csv para los goles que metio Ramsay durante las temporadas entre 2011 y 2014
goles = pd.read_csv('Goles-Ramsay.csv', index_col=0, encoding='latin-1')

#csv para las muertes de famosos que llegaron a ser noticias encabezadas entre 2011 y 2014
muertes = pd.read_csv('Muertes-Temporada.csv', index_col=0, encoding='latin-1')
```

```python
goles
```

```python
muertes
```

```python
#limpieza de datos
goles = goles.reset_index()
muertes = muertes.reset_index()
```

```python
goles = goles[['GOLES QUE METÓ AARON RAMSAY DURANTE LA TEMPORADA','MES Y AÑO']]
goles = goles.drop([12])
```

```python
goles['GOLES QUE METÓ AARON RAMSAY DURANTE LA TEMPORADA'] = goles['GOLES QUE METÓ AARON RAMSAY DURANTE LA TEMPORADA'].astype(np.int64)
```

```python
muertes['NÚMERO DE MUERTES DE FAMOSOS QUE LLEGARON A HACER ENCABEZADOS'] = muertes['NÚMERO DE MUERTES DE FAMOSOS QUE LLEGARON A HACER ENCABEZADOS'].astype(np.int64)
```

```python
goles
```

```python
muertes
```

```python
#plotting
fig, ax1 = plt.subplots() # Generas el primer subplot aquí

ax1.plot_date(goles['MES Y AÑO'], goles['GOLES QUE METÓ AARON RAMSAY DURANTE LA TEMPORADA'], '-', color='red')

plt.xticks(rotation=25)
ax1.legend(['Goles que metió Aaron Ramsay'], fontsize=6)
ax1.set_ylabel('Goles que metió Aaron Ramsay')

ax2 = ax1.twinx() # Creas un nuevo subplot que comparta el mismo eje X
ax2.plot_date(muertes['MES Y AÑO'], muertes['NÚMERO DE MUERTES DE FAMOSOS QUE LLEGARON A HACER ENCABEZADOS'], '--', color='green') # Diferenciamos con línea punteada
ax2.legend(['Número de muertes de famosos que llegaron a hacer encabezados'], fontsize=6, loc=1)
ax2.set_ylabel('Número de muertes de famosos que llegaron a hacer encabezados')

plt.show()
```
